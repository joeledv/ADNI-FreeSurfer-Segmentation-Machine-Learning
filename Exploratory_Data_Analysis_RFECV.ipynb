## High Correlation

```python
threshold = 0.85  # Correlation threshold (adjustable)
df_numeric = df.select_dtypes(include=[np.number])  # Only numeric columns

# 1. IDENTICAL COLUMN DETECTION (CORR = 1)
duplicates_to_drop = set()
for col1, col2 in combinations(df_numeric.columns, 2):
    if df_numeric[col1].equals(df_numeric[col2]):
        print(f"Identical columns found: {col1} == {col2}")
        # We keep the one that appears first in the DataFrame
        duplicates_to_drop.add(col2)

# Immediately remove exact duplicates
df_clean = df_numeric.drop(columns=duplicates_to_drop)

# 2. CORRELATION CALCULATION AND CLUSTERING
corr_matrix = df_clean.corr().abs()
clusters = defaultdict(list)
visited = set()

for col in corr_matrix.columns:
    if col not in visited:
        correlacionadas = corr_matrix.index[corr_matrix[col] > threshold].tolist()

        # Additional verification for near-perfect correlation
        if len(correlacionadas) > 1 and any(corr_matrix.loc[col, c] >= 0.999 for c in correlacionadas[1:]):
            print(f"\nGroup with near-perfect correlation (>0.999): {correlacionadas}")
            # You can add special logic for these cases here

        clusters[len(clusters)] = correlacionadas
        visited.update(correlacionadas)

# 3. REPRESENTATIVE SELECTION (HIGHEST VARIANCE)
columns_to_keep = []
columns_to_drop = list(duplicates_to_drop)  # Include exact duplicates

for cluster in clusters.values():
    if len(cluster) > 1:
        representative = max(cluster, key=lambda x: df_clean[x].var())
        columns_to_keep.append(representative)
        columns_to_drop.extend([x for x in cluster if x != representative])
    else:
        columns_to_keep.extend(cluster)  # Columns without high correlation

# 4. FINAL FILTERING AND REPORT
df_final = df_clean[columns_to_keep]

# Detailed report
print("\n" + "â"*60)
print("CORRELATION REDUCTION SUMMARY")
print("â"*60)
print(f"â¢ Original columns: {len(df_numeric.columns)}")
print(f"â¢ Columns kept: {len(columns_to_keep)}")
print(f"â¢ Columns removed: {len(columns_to_drop)}")
print(f"â¢ Dimensionality reduction: {100*(1-len(columns_to_keep)/len(df_numeric.columns)):.1f}%")

print("\nDETAILS OF REMOVED COLUMNS:")
print("â"*60)
print("1. Identical columns (correlation = 1.0):")
print([col for col in duplicates_to_drop])

print("\n2. Highly correlated columns (>{}):".format(threshold))
for i, cluster in clusters.items():
    if len(cluster) > 1:
        print(f"\n   Group {i+1}:")
        print(f"   â¢ Kept: {max(cluster, key=lambda x: df_clean[x].var())}")
        print(f"   â¢ Removed: {[x for x in cluster if x != max(cluster, key=lambda x: df_clean[x].var())]}")

print("\n" + "â"*60)
print("FINAL DATAFRAME READY FOR MODELING")
print("â"*60)
print("Columns kept (sorted by descending variance):")
print(df_final.var().sort_values(ascending=False).index.tolist())
```

## Feature Scaling

```python
X = df.drop(["Target"], axis=1)
y = df["Target"]
```

```python
feature_names = X.columns.tolist()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42, shuffle=True)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## RFECV implementation

```python
estimator = LogisticRegression(
    solver='liblinear',
    max_iter=1000,
    class_weight='balanced',
    penalty='l2',
    random_state=42,
    C=0.01
)

rfecv = RFECV(
    estimator=estimator,
    step=1,
    cv=3,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=2
)

rfecv.fit(X_train_scaled, y_train)
```

## Important Features

```python
selected_features = [feature_names[i] for i, selected in enumerate(rfecv.support_) if selected]
print(*selected_features, sep="\n")
print(f"\nNumber of selected features: {len(selected_features)}")
```

```python
df= df[selected_features + ['Target']]
```
